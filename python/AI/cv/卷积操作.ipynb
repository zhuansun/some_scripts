```python
import cv2
import numpy as np
# cv2在jupyter中不能展示图像，所以使用maptplotlib
from matplotlib import pyplot as plt
```

```python
# 读取原始图像
img = cv2.imread('beauty.jpg')
img_r = img[:,:,0]
```

```python
# 提取上下不同的卷积核（人为构造卷积核）
kernel = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
```

```python
# 滤波：对图像的一切操作都叫做滤波
img_r_1 = cv2.filter2D(src=img_r, ddepth=-1, kernel=kernel)
```

```python
# 查看原始图像
plt.imshow(X=img_r,cmap='gray')
```

```python
# 查看滤波后的图像
plt.imshow(X=img_r_1,cmap='gray')
```
