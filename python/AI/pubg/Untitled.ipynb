```python
import seaborn as sns
import matplotlib.pyplot as plt

# 假设 train 是一个 Pandas DataFrame，playersJoined 是其中一列
# 以下是示例数据
import pandas as pd
data = {'playersJoined': [10, 20, 10, 30, 20, 10, 40, 30, 20, 10, 50]}
train = pd.DataFrame(data)

# 使用 Seaborn 绘制直方图
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='playersJoined', data=train)


# 添加标题和网格线
plt.title('Number of Players Joined per Game')
plt.xlabel('Number of Players Joined')
plt.ylabel('Count')
plt.grid(True)

# 显示图形
plt.show()
```
